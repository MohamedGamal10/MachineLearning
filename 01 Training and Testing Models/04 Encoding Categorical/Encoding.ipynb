# Onehot encoding

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

# Define dataframe's function and the column to return a dataframe later OHE
def OneHotEncoding(df, column):
    label = LabelEncoder()
    encoding=label.fit_transform(df[column]).reshape(-1,1)
    encoder = OneHotEncoder(sparse=False)
    column = [column+ "_"+ str(i) for i in label.classes_]
    data = encoder.fit_transform(encoding)
    return(pd.DataFrame(data,columns =column))
```

```python
#split categorical data from numerical date
types = pd.DataFrame(dataset.dtypes)
numeric_columns = list(set(types.index[types[0] =="int64"].values) - set(["SalePrice"]))

#create dataframe with ready features
all_data_finish = dataset[numeric_columns]

#create output dataframe
objetive = dataset["SalePrice"]

all_data_finish
```
